```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Stu_id': range(1,101),
    'age':np.random.randint(18,30,100),
    'hours':np.random.randint( 10 , 40,100),
    'attendance':np.random.randint(60,100,100),
    'previous_grade':np.random.randint(70,100,100),
    'extracurriculam':np.random.choice([0,1],100),
    'target':np.random.choice([0,1],100)
}

# Isme begin, end, number of entries
```

```python
# data
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv("Detailss.csv",index=False)
```

```python
df = pd.read_csv('Detailss.csv')
```

```python
df
```

```python
df.shape
```

```python
x = df.drop('target',axis=1)
```

```python
y=df['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=49)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import Perceptron
```

```python
perceptron = Perceptron()
perceptron
```

```python
perceptron.fit(x_train , y_train)
```

```python
y_pred = perceptron.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test , y_pred)
acc
```

```python

# Define the parameter grid for the Perceptron
param_grid = {
    'penalty': ['l2', 'l1', 'elasticnet'],
    'alpha': [0.0001, 0.001, 0.01, 0.1, 1.0],
    'max_iter': [1000, 2000, 3000]
}

# Initialize GridSearchCV with the Perceptron model and parameter grid
grid_search = GridSearchCV(Perceptron(), param_grid, cv=5, scoring='accuracy')

# Fit the GridSearchCV to the training data
grid_search.fit(x_train, y_train)

# Retrieve the best parameters and the best model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

print("Best Parameters:", best_params)

# Evaluate the best model on the test data
y_pred = best_model.predict(x_test)
acc = accuracy_score(y_test, y_pred)

print("Accuracy of the best model:", acc)
```
