```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Stu_id': range(1,101),
    'age':np.random.randint(18,30,100),
    'hours':np.random.randint( 10 , 40,100),
    'attendance':np.random.randint(60,100,100),
    'previous_grade':np.random.randint(70,100,100),
    'extracurriculam':np.random.choice([0,1],100),
    'target':np.random.choice([0,1],100)
}

# Isme begin, end, number of entries
```

```python
# data
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv("Detailss.csv",index=False)
```

```python
df = pd.read_csv('Detailss.csv')
```

```python
df
```

```python
df.shape
```

```python
x = df.drop('target',axis=1)
x
```

```python
y=df['target']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=49)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import Perceptron
```

```python
perceptron = Perceptron()
perceptron
```

```python
perceptron.fit(x_train , y_train)
```

```python
y_pred = perceptron.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test , y_pred)
acc
```

## GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
n_estimators = [20,60,100,120] # number of trees 

max_features = [0.2,0.6,1.0] # number of features consider at every split ( columns )

max_depth = [2,8,None] #  maximum number of levels in tree 

max_samples = [0.5,0.75,1.0] # number of samples 


# 108 (4x3x3x3) random forest train 
```

```python
param_grid = {

    'n_estimators':n_estimators,
    'max_features':max_features,
    'max_depth':max_depth,
    'max_samples':max_samples
}
print(param_grid)
```

```python
Perceptron_grid = GridSearchCV(estimator=perceptron, 
                       param_grid=param_grid,
                       cv=5,
                       verbose=2,
                       n_jobs= -1)
```
