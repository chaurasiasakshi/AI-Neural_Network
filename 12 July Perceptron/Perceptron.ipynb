```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Stu_id': range(1,101),
    'age':np.random.randint(18,30,100),
    'hours':np.random.randint( 10 , 40,100),
    'attendance':np.random.randint(60,100,100),
    'previous_grade':np.random.randint(70,100,100),
    'extracurriculam':np.random.choice([0,1],100),
    'target':np.random.choice([0,1],100)
}

# Isme begin, end, number of entries
```

```python
# data
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv("Detailss.csv",index=False)
```

```python
df = pd.read_csv('Detailss.csv')
```

```python
df
```

```python
df.shape
```

```python
x = df.drop('target',axis=1)
x
```

```python
y=df['target']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=49)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import Perceptron
```

```python
perceptron = Perceptron()
perceptron
```

```python
perceptron.fit(x_train , y_train)
```

```python
y_pred = perceptron.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test , y_pred)
acc
```

## GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
perceptron_grid = GridSearchCV(Perceptron(), {
    'penalty': ['l2', 'l1', 'elasticnet'],
    'alpha': [0.0001, 0.001, 0.01, 0.1]
}, cv=5, return_train_score=False)

perceptron_grid.fit(x_train, y_train)
results = perceptron_grid.cv_results_
# print(results)
```

```python
df = pd.DataFrame(results)
print(df)
```

```python
columns = df[['param_alpha', 'mean_test_score', 'rank_test_score']]
print(columns)
```

```python
# GridSearchCV
```

```python
param_grid = {

    'penalty': ['l2', 'elasticnet'],  # Regularization terms

    'alpha': [0.0001, 0.001, 0.01, 0.1,2.0],  # Learning rate

    'max_iter': [1000, 2000, 3000]  # Number of iterations

}
```

```python
# Perform GridSearchCV to find the best hyperparameters
grid_search = GridSearchCV(perceptron, param_grid, cv=15)
grid_search.fit(x_train, y_train)
```

```python
best_perceptron = grid_search.best_estimator_
best_perceptron.fit(x_train, y_train)
```

```python
y_pred = best_perceptron.predict(x_test)
 
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy of the improved Perceptron model: {accuracy:.2f}')
```
